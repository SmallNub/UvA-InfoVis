```python
import plotly.graph_objs as go
import plotly.express as px
import pandas as pd
from plotly import colors
```

- Import needed libaries


```python
df = pd.read_csv('riaakaggle.csv') # import file
df = df.head(n=50) # take first 50 entries
df

```

- File is sortend on certified units, take first 50

```python
x_axis = df['Artist']

data = [
    go.bar(x=x_axis, y=df['Certified Units']),
    go.bar(x=x_axis, y=df['Gold']),
    go.bar(x=x_axis, y=df['Platinum']),
    go.bar(x=x_axis, y=df['Multi-Platinum']),
    go.bar(x=x_axis, y=df['Diamond'])
]

layout = go.Layout(
    title='Percentage unmarried in the US between 25 and 34 by income',
    height=400,
    xaxis=go.layout.XAxis(
        type='category', # the type for the x axis is categorical
        title='Artist',    # x-axis title
        tickangle=0,    # change label to horizontal
        tickfont=dict(size=10) # change label size
    ),
    yaxis=go.layout.YAxis(
        tickformat=',.0%', # show as percentage
        title='Albums' # y-axis title
    )
)

fig = go.Figure(data=data, layout=layout)
fig.show()
```
